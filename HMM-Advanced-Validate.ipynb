```python
# Plot first IT avg with error bars vs n clusters
ys = vd.hmm_its[:,:,0,:4]
ysavg = np.average(ys, axis=0)
yserr = np.std(ys, axis=0)
xs = vd.params[1]
for i in xrange(ysavg.shape[1]):
    pp.errorbar(xs,ysavg[:,i], yerr=yserr[:,i],label='it %d'%(i+1))
#pp.hlines(vd.analytic_its[:4], 0,20)
pp.yscale('log')
pp.ylim(1e2, 1e5)
pp.xlabel('Number of clusters')
pp.title('HMM Implied Timescales vs. k')
pp.legend(loc='lower right', ncol=2)
pp.show()
```

```python
# Plot first IT avg with error bars vs n clusters
ys = vd.msm_its[:,:,0,:4]
ysavg = np.average(ys, axis=0)
yserr = np.std(ys, axis=0)
xs = vd.params[1]
for i in xrange(ysavg.shape[1]):
    pp.errorbar(xs,ysavg[:,i], yerr=yserr[:,i],label='it %d'%(i+1))
#pp.hlines(vd.analytic_its[:4], 0,20)
pp.yscale('log')
pp.xlabel('Number of clusters')
pp.title('MSM Implied Timescales vs. k')
pp.legend(loc='lower right', ncol=2)
pp.ylim(1e2,1e5)
pp.show()
```

```python
# Plot first IT avg with error bars vs n clusters
def plots_vs_k(it_i, lt_i, ax=None):
    xs = vd.params[1]
    
    if ax is None:
        ax = pp
    
    # HMM
    ys = vd.hmm_its[:,:,lt_i,it_i]
    ysavg = np.average(ys, axis=0)
    yserr = np.std(ys, axis=0)
    ax.errorbar(xs,ysavg, yerr=yserr,label='hmm')
    
    # MSM
    ys = vd.msm_its[:,:,lt_i,it_i]
    ysavg = np.average(ys, axis=0)
    yserr = np.std(ys, axis=0)
    ax.errorbar(xs,ysavg, yerr=yserr,label='msm')
    
    ax.hlines(vd.analytic_its[it_i], 0,20)
    ax.set_yscale('log')
    ax.legend()
    ax.set_title("Implied Timescale %d, lt = %d"%(it_i+1, vd.params[2][lt_i]))
    ax.set_xlabel('Number of clusters')
```

```python
# vs k
fig, axes = pp.subplots(nrows=3, ncols=2, figsize=(12,12))
for row in range(3):
    for col in range(2):
        plots_vs_k(row, [0,5][col], axes[row,col])
fig.tight_layout()
fig.savefig('its_vs_k.png')
```

```python
# Plot first IT avg with error bars vs lag time
def plots_vs_lt(it_i, k_i, ax):
    xs = vd.params[2]
    
    # HMM
    ys = vd.hmm_its[:,k_i,:,it_i]
    ysavg = np.average(ys, axis=0)
    yserr = np.std(ys, axis=0)
    ysavg = np.clip(ysavg, 0, 1e6)
    yserr = np.clip(yserr, 0, 1e4)
    ax.errorbar(xs,ysavg, yerr=yserr,label='hmm')
    
    # MSM
    ys = vd.msm_its[:,k_i,:,it_i]
    ysavg = np.average(ys, axis=0)
    yserr = np.std(ys, axis=0)
    ax.errorbar(xs,ysavg, yerr=yserr,label='msm')
    
    ax.hlines(vd.analytic_its[it_i], 0,30)
    ax.set_yscale('log')
    ax.legend()
    ax.set_xlabel('Lag Time')
    ax.set_title("Implied Timescale %d, k = %d"%(it_i+1, vd.params[1][k_i]))
```

```python
# vs. lt
fig, axes = pp.subplots(nrows=3, ncols=2, figsize=(12,12))
for row in range(3):
    for col in range(2):
        plots_vs_lt(row, [1,2][col], axes[row,col])
fig.tight_layout()
fig.savefig('its_vs_lt.png')
```
