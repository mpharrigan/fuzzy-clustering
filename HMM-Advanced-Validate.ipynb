Using HMMs on the Muller potential
=========

This notebook uses an analytic transition matrix to compare performance and accuracy between approximate markov methods. A high resolution transition matrix is loaded and used to generate 'trajectories' to which we will apply our methods. In this notebook, convergence of implied timescales is compared for the HMM scheme vs a traditional discrete-state MSM in relation to number of states as well as lag-time. I hypothesize that the HMM model will be more accurate with fewer states and a lower lag time

## Functions

```python
def compute_its(tmatrix_func, analysis_func, param_range, n_timescales):
    implied_timescales = np.zeros((0, n_timescales + 1))
    for param in param_range:
        _tmatrix = tmatrix_func(param)
        _it = analysis_func(_tmatrix, param, n_timescales)
        implied_timescales = np.append(implied_timescales, [np.append(param, _it)], axis=0)
        print "Computed param", param
    return implied_timescales
```

```python
def _plot_it(its, it_i, label):
    ys = list()
    xs = list()
    label = label + (" IT %d" % it_i)
    for it in its:        
        ys.append(it[it_i])
        xs.append(it[0])
    pp.plot(xs, ys, 'o-', label=label)  
    

def plot_it(its, it_i, labels=None, bic=None):
    """Plot a specified implied timescale."""
    
    if isinstance(it_i, list):
        pass
    elif isinstance(it_i, int):
        it_i = [it_i]
    else:
        raise ValueError("IT index must be an int or list of ints")
        
    if isinstance(its, list):
        pass
    elif isinstance(its, np.ndarray):
        its = [its]
    else:
        raise ValueError("Implied timescales must be an ndarray or list of ndarrays.")
        
    if isinstance(labels, list):
        pass
    elif isinstance(labels, str):
        labels = [labels]
        
    
    if labels is None:
        labels = ['Scheme %d' % (i+1) for i in xrange(len(its))]
    
    assert len(labels) == len(its), 'Number of labels must match number of things to plot.'        
        
    for i in it_i:
        for j in xrange(len(its)):
            _plot_it(its[j], i, labels[j])

    pp.yscale('log')
    
    for i in it_i:
        # Plot analytic
        xmin, xmax = pp.xlim()
        pp.hlines(analytic_it[i - 1], xmin, xmax)
    

    
    # Optionally highlight best BIC
    if bic is not None:
        ymin, ymax = pp.ylim()
        pp.vlines(bic, ymin, ymax)
    
    pp.legend(loc='lower left')
    pp.show()
    
    
```

```python
# Compute HMM by number of clusters
def hmm_tmat_k(k): return mixture.hmm(traj_list, fix_k=k, lag_time=1, sliding_window=True)[0]
def hmm_anal_k(tmatrix, k, n_timescales): return analysis.get_implied_timescales(tmatrix, n_timescales, lag_time=stride*1)
hmm_its_k = compute_its(hmm_tmat_k, hmm_anal_k, xrange(2,15), n_timescales=6)
```

## Lag-times

```python
# Compute MSM implied timescales vs lag-time
msm_its_lt = np.zeros((0,max_timescale))
for lt in xrange(1,20):
    msm_tmatrix = classic.msm(traj_list, n_clusters=80, n_medoid_iters=10, lag_time=lt)
    msm_it = analysis.get_implied_timescales(msm_tmatrix, n_timescales=max_timescale, lag_time=stride*lt)
    msm_its_lt = np.append(msm_its_lt, [msm_it], axis=0)
    print "Computed lt = %d" % lt
```

```python
# Compute HMM implied timescales vs lag-time
hmm_its_lt = np.zeros((0,max_timescale))
for lt in xrange(1,20):
    hmm_tmatrix, _ = mixture.hmm(traj_list, fix_k=10, lag_time=lt, sliding_window=True)
    hmm_it = analysis.get_implied_timescales(hmm_tmatrix, n_timescales=max_timescale, lag_time=stride*lt)
    hmm_its_lt = np.append(hmm_its_lt, [hmm_it], axis=0)
    print "Computed lt = %d" % lt
```
