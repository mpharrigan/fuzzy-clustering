Using HMMs on the Muller Potential
==============
This notebook validates the HMM procedure and shows that---on the muller potential---HMMs can perform better than a traditional, identity-function-basis-set MSM in terms of matching Implied Timescales.

In this example, a high-resolution (32600 state) transition matrix is generated analytically from the Muller potential. Trajectories are generated from this transition matrix. The two schemes are applied to these trajectories, and we can compare the approximate implied timescales to the analytical ones obtained from the actual generating transition matrix.

Code for (sort-of) extracting eigenfunctions
-----------
Note: actual eigenfunctions are probably more complicated

```python
def multi_norm(point, mu, sigma):
    k = len(point)
    assert k == len(mu)
    assert k == sigma.shape[0]
    assert k == sigma.shape[1]
    
    norm = 1.0 / np.sqrt((2 * np.pi)**k * linalg.det(sigma))
    exp = -0.5 * np.dot((point - mu).T,  np.dot(linalg.inv(sigma), (point - mu)))
    return norm * np.exp(exp)

def sum_multi_norm(point, means, weights, covars):
    n_mix = len(means)
    res = 0.0
    for mean, weight, covar in zip(means,weights,covars):
        res += weight * multi_norm(point, mean, covar)
    return res

def mix(points, means, weights, covars):
    z = np.zeros(len(points))
    for i, point in enumerate(points):
        z[i] = sum_multi_norm(point, means, weights, covars)
    return z
```
