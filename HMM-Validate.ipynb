Using HMMs on the Muller Potential
==============
This notebook validates the HMM procedure and shows that---on the muller potential---HMMs can perform better than a traditional, identity-function-basis-set MSM in terms of matching Implied Timescales.

In this example, a high-resolution (32600 state) transition matrix is generated analytically from the Muller potential. Trajectories are generated from this transition matrix. The two schemes are applied to these trajectories, and we can compare the approximate implied timescales to the analytical ones obtained from the actual generating transition matrix.

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
from __future__ import division
from fuzzy import mixture, classic, analysis
from msmtoys import plotting, muller
import msmtoys.analytic
```

```python
from matplotlib import pyplot as pp
```

Generate a transition matrix
---------
The transition matrix is generated at a certain resolution and temperature ~ 1/beta

```python
# Optional: Generate a transition matrix
import pickle
t_matrix, gird = msmtoys.analytic.calculate_transition_matrix(muller.MullerForce, resolution=200, beta=1/70)
with open('transition_matrix_70.pickl', 'wb') as f:
    pickle.dump((t_matrix,grid), f)
```

```python
# Or: Load from file
import pickle
with open('transition_matrix.pickl', 'rb') as f:
    t_matrix, grid = pickle.load(f)
print t_matrix.shape
```

Generate Trajectories
----------------
We generate a number of trajectories from the transition matrix

```python
# Generate and view trajectories
num_trajs = 2
traj_len = 100000
stride = 10
print "Aggregate: {:,} points".format((num_trajs*traj_len/stride))
traj_list = msmtoys.analytic.get_trajlist(t_matrix, grid, num_trajs, traj_len, stride)
plotting.plot2d(muller.MullerForce)
[pp.plot(traj[:,0], traj[:,1], '-') for traj in traj_list]
pp.title("Sample trajectories")
pp.show()
```

```python
# Use HMM Model
# Pick k based on BIC
hmm_tmatrix, mixture_model = mixture.hmm(traj_list, lag_time=30)
hmm_tmatrix.shape
```

```python
# Look at the mixture model
mixture.plot_distribution(mixture_model, grid)
pp.title("Mixture components")
pp.show()
```

```python
# Use classic MSM with a small number of clusters
smallk_tmatrix = classic.msm(traj_list, n_clusters=9, n_medoid_iters=10, lag_time=1)
smallk_tmatrix.shape
```

```python
# Use classic MSM with a large number of clusters
bigk_tmatrix = classic.msm(traj_list, n_clusters=200, n_medoid_iters=10, lag_time=1)
bigk_tmatrix.shape
```

```python
# This takes a while
n=5
analytic_it = analysis.get_implied_timescales(t_matrix, n_timescales=n, lag_time=1)
```

```python
hmm_it = analysis.get_implied_timescales(hmm_tmatrix, n_timescales=n, lag_time=stride*30)
smallk_it = analysis.get_implied_timescales(smallk_tmatrix, n_timescales=n, lag_time=stride)
bigk_it = analysis.get_implied_timescales(bigk_tmatrix, n_timescales=n, lag_time=stride)
```

```python
its = [analytic_it, hmm_it, smallk_it, bigk_it]
descs = ['Analytic', 'HMM', 'Classic (small k)', 'Classic (big k)']
analysis.plot_lambda_bar(its, descs, logplot=True)
```
