```python
from __future__ import division
from fuzzy import mixture, classic, analysis
from msmtoys import plotting, muller
import msmtoys.analytic
```

```python
from matplotlib import pyplot as pp
```

```python
# Optional: Generate a transition matrix
import pickle
t_matrix, gird = msmtoys.analytic.calculate_transition_matrix(muller.MullerForce, resolution=200, beta=1/30)
with open('transition_matrix.pickl', 'wb') as f:
    pickle.dump((t_matrix,grid), f)
```

```python
# Or: Load from file
import pickle
with open('transition_matrix.pickl', 'rb') as f:
    t_matrix, grid = pickle.load(f)
print t_matrix.shape
```

```python
# Generate and view trajectories
num_trajs = 100
traj_len = 2000
stride = 10
print "Aggregate: {:,} points".format((num_trajs*traj_len/stride))
```

```python
traj_list = msmtoys.analytic.get_trajlist(t_matrix, grid, num_trajs, traj_len, stride)
plotting.plot2d(muller.MullerForce)
[pp.plot(traj[:,0], traj[:,1], '-') for traj in traj_list]
pp.title("Sample trajectories")
pp.show()
```

```python
# Use HMM Model
# Pick k based on BIC
hmm_tmatrix, mixture_model = mixture.hmm(traj_list, lag_time=1)
hmm_tmatrix.shape
```

```python
# Look at the mixture model
#plotting.plot2d(muller.MullerForce)
mixture.plot_distribution(mixture_model, grid)
pp.title("Mixture components")
pp.show()
```

```python
# Use classic MSM with a small number of clusters
smallk_tmatrix = classic.msm(traj_list, n_clusters=11, n_medoid_iters=10, lag_time=1)
smallk_tmatrix.shape
```

```python
# Use classic MSM with a large number of clusters and a higher lag time
bigk_tmatrix = classic.msm(traj_list, n_clusters=200, n_medoid_iters=10, lag_time=1)
bigk_tmatrix.shape
```

```python
# Use classic MSM with a large number of clusters and the same lag time
bigkhighlt_tmatrix = classic.msm(traj_list, n_clusters=200, n_medoid_iters=10, lag_time=10)
bigkhighlt_tmatrix.shape
```

```python
# This takes a while
n=5
analytic_it = analysis.get_implied_timescales(t_matrix, n_timescales=n, lag_time=1)
```

```python
hmm_it = analysis.get_implied_timescales(hmm_tmatrix, n_timescales=n, lag_time=stride)
smallk_it = analysis.get_implied_timescales(smallk_tmatrix, n_timescales=n, lag_time=stride)
bigk_it = analysis.get_implied_timescales(bigk_tmatrix, n_timescales=n, lag_time=stride)
bigkhighlt_it = analysis.get_implied_timescales(bigk_tmatrix, n_timescales=n, lag_time=10*stride)
```

```python
its = [analytic_it, hmm_it, smallk_it, bigk_it, bigkhighlt_it]
descs = ['Analytic', 'HMM', 'Classic (small k)', 'Classic (big k)', 'Classic (big k, high lt)']
analysis.plot_lambda_bar(its, descs, logplot=True)
```
